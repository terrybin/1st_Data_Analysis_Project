```python
# (본론 Fig.2) 서울 도시철도 연간 재정 추이 데이터 제작 및 확인 (참조: 서울시 교통기획관 교통정책과)
seoulTrainLoss = pd.read_excel('./Fixed Data/seoulTrainLoss.xlsx', engine='openpyxl')
seoulTrainLoss.columns = ['연도', '운영비용', '영업수익', '적자']
seoulTrainLoss
```

상기 그래프를 통해 서울 도시철도의 운영비용은 계속 증가하는 데 반해 영업수익이 저조함을 알 수 있음  
또한 적자가 2022년 1조 2천억까지 증가하면서 도시철도의 적자 개선이 시급한 상황  
적자가 계속 증가하는 원인이 무임승차 인원이라고 가정하고 검증을 위해 2022년 연간 유/무임 승차 인원 파악 진행
***

```python
# (본론 Fig.3) 연도별 유무임 승차인원 및 비율 데이터 가공 (데이터: 서울 도시철도 1 ~ 9호선)
rideCount_by_fare = pd.read_csv('./Fixed Data/seoulTrain_yearlyRide_ver.2.csv', encoding='euc-kr')
rideCount_by_fare
```

```python
# 그래프로 대조하기 위해 무임승차비율 drop
fixedData_rideCount = rideCount_by_fare.drop(columns=['무임승차비율(%)'])
fixedData_rideCount
```

```python
# 단위 백만명 기준 columns 추가
fixedData_rideCount['총계(백만명)'] = fixedData_rideCount['총계(명)'] / 1000000
fixedData_rideCount['유임승차인원(백만명)'] = fixedData_rideCount['유임승차인원(명)'] / 1000000
fixedData_rideCount['무임승차인원(백만명)'] = fixedData_rideCount['무임승차인원(명)'] / 1000000
fixedData_rideCount
```

```python
# (본론 Fig.3) 연도별 도시철도 유/무임 이용고객 현황
plt.figure(figsize=(12, 8))
plt.plot(fixedData_rideCount['연도'], fixedData_rideCount['총계(백만명)'], label='총계')
plt.plot(fixedData_rideCount['연도'], fixedData_rideCount['유임승차인원(백만명)'], label='유임승차인원', color='green')
plt.plot(fixedData_rideCount['연도'], fixedData_rideCount['무임승차인원(백만명)'], label='무임승차인원', color='red')
plt.yticks(np.arange(250, 3250, 250))
plt.title('2016 ~ 2022년 연도별 유/무임 이용고객 현황 (단위: 백만명)')
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 15

plt.legend()
plt.show()
```

상기 그래프로 확인했을 때 무임승차인원의 비율은 2016년부터 7년간 약 2억명 수준의 인원 감소세를 보임  
반면 총 승차인원과 유임승차인원의 경우 2016년부터 지속적으로 감소하여 7년간 약 12 ~ 14억명 수준의 인원 감소세를 보임  
즉, 도시철도를 이용하는 인원이 줄어듦에 따라 2022년 총 인원 대비 무임승차비율은 2016년에 비해 약 1.3% 정도 상승

```python
# (본론 Fig.4) 장래인구 추계 데이터 Read (2022 ~ 2030)
future_personRatio = pd.read_excel('./Fixed Data/future_peopleChange_ratio_ver.2.xlsx', engine='openpyxl')
future_personRatio
```

- 총부양비=유소년부양비+노년부양비
- 유소년부양비=유소년인구/생산연령인구 * 100
- 노년부양비=고령인구/생산연령인구 * 100
- 노령화지수=고령인구/유소년인구 * 100

```python
# (본론 Fig.4) 장래인구 부양비 추이
# 2022 ~ 2030년 총 부양비 및 유소년/노년 부양비 추이 확인
plt.figure(figsize=(12, 8))
plt.plot(future_personRatio['연도'], future_personRatio['총부양비'], label='총 부양비(%)')
plt.plot(future_personRatio['연도'], future_personRatio['유소년부양비'], label='유소년부양비(0 ~ 14세, %)')
plt.plot(future_personRatio['연도'], future_personRatio['노년부양비'], label='노년부양비(65세 이상, %)')

plt.title('2022 ~ 2030 장래인구 부양비 추이')
plt.legend()
plt.show()
```

승차 피크시간대 확인을 위한 2022년 월별, 시간대별, 호선별 승차인원 파악 

```python
# 결론 제시 중 노인 무임승차제도의 개정을 위해 국내 도시철도의 승차 피크시간대 확인 필요(대상: 1 ~ 9 호선, 기간: 2022년 12월)
trainRideCount = pd.read_excel('./Fixed Data/trainRideCount_year_ver.2.xlsx', engine='openpyxl')
trainRideCount
```

```python
# 'T' 속성을 지정해서 describe() 함수 실행 결과를 전치시켜 요약 통계량을 수직 방향으로 출력
trainRideCount.groupby(trainRideCount['호선명']).describe().T
```

```python
# 피벗 테이블화 (2022년 1 ~ 9호선 시간대별 승차인원 확인)
rideCountPivot = trainRideCount.pivot_table('승차인원', '승차시간 구분', '호선명', 'min')
rideCountPivot
```

```python
# 시간대별 승차인원 현황 피벗테이블을 히트맵으로 시각화
sns.heatmap(rideCountPivot, annot_kws={'size:20'}, cmap='PuRd').set_title(
    '2022년 12월 승차인원 승차시간 파악', fontdict = {'fontsize': 20, 'fontfamily': 'NanumGothicCoding'}
)
sns.set(rc = {'figure.figsize':(22,15)})
plt.rc('font', family='NanumGothicCoding')
plt.legend(np.arange(0, 6000000, 300000))
```

상단 히트맵을 통해 각 호선별로 07 ~ 10시 승차와 16 ~ 20시 승차 두 타임이 가장 승차인원이 많은 것으로 확인  
따라서 노인 무임승차제도를 피크시간에 요금을 부과하는 방향으로 개정한다면 07 ~ 10시 / 16 ~ 20시 두 타임에 대해  
일반 이용고객들과 동일한 요금을 부과하도록 하는 것이 서울 도시철도의 적자개선에 도움이 될 것으로 보임
